```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.barttorvik.com/"
response = requests.get(url)

# Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")

# Inspect the page structure and find the table or relevant data
# Example: Extracting table data
table = soup.find("table")
rows = table.find_all("tr")

data = []
for row in rows:
    cols = row.find_all("td")
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)

# Convert to DataFrame and save as CSV
df = pd.DataFrame(data)
df = df.dropna(how="all").reset_index(drop=True)

df.columns = ['RK', 'TEAM', 'CONF', 'G', 'REC', 'ADJOE', 'ADJDE', 'BARTHAG', "EFG_O", "EFG_D", 'TOR', 'TORD', 'ORB', 'DRB', 'FTR', 'FTRD', "2P_O", "2P_D", "3P_O", "3P_D", '3PR', '3PRD', 'ADJ_T', 'WAB']
df = df[[
    "TEAM", "CONF", "G", "ADJOE", "ADJDE", "BARTHAG", 
    "EFG_O", "EFG_D", "TOR", "TORD", "ORB", "DRB", "FTR", "FTRD", 
    "2P_O", "2P_D", "3P_O", "3P_D", "ADJ_T", "WAB"
]]
df.to_csv("current_cbb_live_data.csv", index=False)
```

```python
from pygam import LinearGAM, s

# Replace NaN values in train and test datasets
df_train = cbb.replace(np.nan, 0)
df_test = cbb24.replace(np.nan, 0)
```

```python
df_train
```

```python
# Define a mapping dictionary for POSTSEASON finishes
postseason_mapping = {
    "Champions": 1,
    "2ND": 2,
    "F4": 3,
    "E8": 4,
    "S16": 5,
    "R32": 6,
    "R64": 7,
    "R68": 8
}

# Apply the mapping to create POSTSEASON_FINISH column
df_train["POSTSEASON"] = df_train["POSTSEASON"].map(postseason_mapping).fillna(0).astype(int)
```

```python
df_train
```

```python

# Define relevant features based on final data structure
selected_features = [
    "G", "ADJOE", "ADJDE", "BARTHAG", "EFG_O", "EFG_D",
    "TOR", "TORD", "ORB", "DRB", "FTR", "FTRD",
    "2P_O", "2P_D", "3P_O", "3P_D", "ADJ_T", "WAB"
]

# Prepare train data
X_train = df_train[selected_features]
y_train = df_train["POSTSEASON"]  # Target variable is now PostSeason
```

```python
# Build and fit a GAM model
gam = LinearGAM(
    s(0) + s(1) + s(2) + s(3) + s(4) + s(5) + s(6) +
    s(7) + s(8) + s(9) + s(10) + s(11) + s(12) + s(13) +
    s(14) + s(15) + s(16)
)
gam.fit(X_train, y_train)
```

```python

# Prepare test data
X_test = df_test[selected_features]
teams_test = df_test["TEAM"]  # Preserve team names for final output

# Predict postseason outcomes for the test data
y_pred_test = gam.predict(X_test)

# Combine predictions with team names
results_test = pd.DataFrame({
    "TEAM": teams_test,
    "Predicted_PostSeason": y_pred_test
})

# Sort results based on predicted postseason finishes
ranked_results_test = results_test.sort_values(
    by="Predicted_PostSeason", ascending=False
).reset_index(drop=True)

# Print ranked teams with predicted postseason finishes
print(ranked_results_test)
```

```python
ranked_results_test['Predicted_PostSeason'] = np.floor(ranked_results_test['Predicted_PostSeason'])

# Filter teams that make March Madness (Predicted_PostSeason > 0.0)
teams_make_march_madness = ranked_results_test[ranked_results_test['Predicted_PostSeason'] > 0.0]
```
